```python
import sqlalchemy
import pandas as pd
from sqlalchemy import create_engine as ce
from sqlalchemy import Column, String, Integer
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
```

```python
#Remember Divide the Python Script to extraction and loading

import requests
import json
from pandas import DataFrame, json_normalize
import datetime as dt
# Root Endpoint
# query data with GET - get all beers as json
```

```python
BASE_URL = "https://api.punkapi.com/v2/"
# query data with GET - get all beers as json
response = requests.get(f'{BASE_URL}/beers')
print(response.status_code)

# > 200
```

```python
# Response's headers
print(response.headers)
 # Get the json
r_string = response.json()
#dump the json result into a string
ex_string = json.dumps(r_string)  
print(ex_string)
```

```python
#load the string into a json object
api_json = json.loads(ex_string)
#Put in DataFrame
d = pd.DataFrame(api_json)

d
```

```python
                       # Transformation Part
#Select specific columns from the Dataframe


dd= pd.DataFrame(d, columns= ['id', 'name', 'first_brewed', 'srm', 'ph', 'abv'])
print(dd)
```

```python
f = ce("sqlite:///BeerAPI.db")
```

Loading Api Data to SQlite Database

```python
dd.to_sql('BierAPI_table', f)
```

Connecting to postgres

```python
pg_engine=ce("postgresql+psycopg2://postgres:2892@localhost:5432/postgres")
```

Loading API Data to Postgress

```python
dd.to_sql("BeerAPI", pg_engine)
```
